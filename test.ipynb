```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://poshmark.com/category/Women-Dresses'  
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
listings = soup.find_all('div', class_='card card--small')

# Create a list to store your data
data = []

for listing in listings:
  title_tag = listing.find('a', class_='tile__title')
  price_tag = listing.find('span', class_='fw--bold')
  size_tag = listing.find('a', class_='tile__details__pipe__size')
  brand_tag = listing.find('a', class_='tile__details__pipe__brand')

  if title_tag and price_tag and size_tag and brand_tag:
    title = title_tag.text.strip()
    price = price_tag.text.strip()
    size = size_tag.text.strip()
    brand = brand_tag.text.strip()

    # Create a dictionary with your data and append it to the list
    data.append({
      'Title': title,
      'Price': price,
      'Size': size,
      'Brand': brand
    })

# Convert the list to a DataFrame
df = pd.DataFrame(data)

# Write the DataFrame to an Excel file
df.to_excel('output.xlsx', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://poshmark.com/category/Women-Dresses?availability=available&sort_by=like_count&all_size=true&my_size=false'  
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
listings = soup.find_all('div', class_='card card--small')


data = []

for listing in listings:
    title_tag = listing.find('a', class_='tile__title')
    price_tag = listing.find('span', class_='fw--bold')
    size_tag = listing.find('a', class_='tile__details__pipe__size')
    brand_tag = listing.find('a', class_='tile__details__pipe__brand')

    if title_tag and price_tag and size_tag and brand_tag:
        title = title_tag.text.strip()
        price = price_tag.text.strip()
        size = size_tag.text.strip()
        brand = brand_tag.text.strip()

        
        data.append({
            'Title': title,
            'Price': price,
            'Size': size,
            'Brand': brand
        })


df = pd.DataFrame(data)
df.to_excel('rating.xlsx', index=False)
```
